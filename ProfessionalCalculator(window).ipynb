```python
# Professional Calculator Project
# Python Tkinter GUI
# Run in Jupyter Notebook

import tkinter as tk
from tkinter import messagebox


class ProfessionalCalculator:

    def __init__(self, root):
        self.root = root

        # Window settings
        self.root.title("Professional Calculator")
        self.root.geometry("420x600")
        self.root.resizable(False, False)

        # Color Theory Theme
        # Dark background = reduces eye strain
        # Blue accent = trust/professional feeling
        # Orange = attention/action buttons
        self.bg_color = "#121826"
        self.display_color = "#1F2937"
        self.button_color = "#243447"
        self.operator_color = "#2563EB"
        self.equal_color = "#F59E0B"
        self.text_color = "#FFFFFF"
        self.hover_color = "#334155"

        self.root.configure(bg=self.bg_color)

        self.expression = ""

        self.create_widgets()

        # Keyboard support
        self.root.bind("<Key>", self.keyboard_input)


    def create_widgets(self):

        # Display
        self.display = tk.Entry(
            self.root,
            font=("Arial", 28, "bold"),
            bg=self.display_color,
            fg=self.text_color,
            bd=0,
            justify="right"
        )

        self.display.pack(
            fill="both",
            padx=20,
            pady=25,
            ipady=20
        )


        # Button Frame
        frame = tk.Frame(
            self.root,
            bg=self.bg_color
        )

        frame.pack(expand=True, fill="both")


        buttons = [

            ["C", "⌫", "%", "/"],
            ["7", "8", "9", "*"],
            ["4", "5", "6", "-"],
            ["1", "2", "3", "+"],
            ["0", ".", "=", ]

        ]


        for r, row in enumerate(buttons):

            for c, btn in enumerate(row):

                if btn == "=":
                    color = self.equal_color

                elif btn in ["+", "-", "*", "/", "%"]:
                    color = self.operator_color

                elif btn in ["C", "⌫"]:
                    color = "#DC2626"

                else:
                    color = self.button_color


                button = tk.Button(
                    frame,
                    text=btn,
                    font=("Arial", 20, "bold"),
                    fg=self.text_color,
                    bg=color,
                    activebackground=self.hover_color,
                    activeforeground="white",
                    bd=0,
                    command=lambda x=btn: self.click(x)
                )

                button.grid(
                    row=r,
                    column=c,
                    sticky="nsew",
                    padx=8,
                    pady=8
                )


        # Grid responsive
        for i in range(5):
            frame.rowconfigure(i, weight=1)

        for i in range(4):
            frame.columnconfigure(i, weight=1)


    def click(self, value):

        if value == "C":

            self.expression = ""
            self.display.delete(0, tk.END)


        elif value == "⌫":

            self.expression = self.expression[:-1]
            self.display.delete(0, tk.END)
            self.display.insert(
                tk.END,
                self.expression
            )


        elif value == "=":

            try:
                result = eval(self.expression)

                self.display.delete(0, tk.END)
                self.display.insert(
                    tk.END,
                    result
                )

                self.expression = str(result)


            except:

                messagebox.showerror(
                    "Error",
                    "Invalid Calculation"
                )

                self.expression = ""


        else:

            self.expression += str(value)

            self.display.delete(
                0,
                tk.END
            )

            self.display.insert(
                tk.END,
                self.expression
            )


    def keyboard_input(self, event):

        key = event.char

        if key in "0123456789+-*/.%":

            self.click(key)


        elif key == "\r":

            self.click("=")


        elif key == "\x08":

            self.click("⌫")



# Start Application

window = tk.Tk()

app = ProfessionalCalculator(window)

window.mainloop()
```
