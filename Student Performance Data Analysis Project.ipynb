```python
# ==========================================
# Student Performance Data Analysis Project
# ==========================================

import pandas as pd
import matplotlib.pyplot as plt


# Create Dataset
data = {
    "Name": ["Ali", "Ahmed", "Sara", "Fatima", "Usman", "Ayesha"],
    "Math": [85, 78, 92, 88, 70, 95],
    "English": [80, 75, 90, 85, 72, 94],
    "Computer": [90, 82, 95, 89, 75, 98]
}


# Convert to DataFrame
df = pd.DataFrame(data)


print("===== Student Data =====")
print(df)


# Calculate Total Marks
df["Total"] = (
    df["Math"] +
    df["English"] +
    df["Computer"]
)


# Calculate Average
df["Average"] = df["Total"] / 3


print("\n===== Student Result =====")
print(df)


# Sort Students by Performance
ranking = df.sort_values(
    by="Average",
    ascending=False
)


print("\n===== Student Ranking =====")
print(ranking)


# Find Best Student
best_student = df.loc[df["Average"].idxmax()]


print("\n===== Top Student =====")
print(
    "Name:",
    best_student["Name"]
)
print(
    "Average Marks:",
    best_student["Average"]
)


# Subject Wise Analysis
subjects = [
    "Math",
    "English",
    "Computer"
]


subject_average = df[subjects].mean()


print("\n===== Subject Average =====")
print(subject_average)


# Chart 1: Subject Average
plt.figure(figsize=(8,5))

subject_average.plot(
    kind="bar"
)

plt.title(
    "Average Marks by Subject"
)

plt.xlabel(
    "Subjects"
)

plt.ylabel(
    "Average Marks"
)

plt.show()



# Chart 2: Student Performance
plt.figure(figsize=(8,5))

df.plot(
    x="Name",
    y="Average",
    kind="bar",
    figsize=(8,5)
)

plt.title(
    "Student Average Performance"
)

plt.xlabel(
    "Students"
)

plt.ylabel(
    "Average Marks"
)

plt.show()


print("\nProject Completed Successfully ✅")
```
