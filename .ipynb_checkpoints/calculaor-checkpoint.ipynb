```python
# ==========================================
# Advanced Calculator in Python
# Jupyter Notebook Project
# ==========================================

import math

def calculator():
    print("="*40)
    print("       PYTHON ADVANCED CALCULATOR")
    print("="*40)

    while True:
        print("\nSelect Operation:")
        print("1. Addition (+)")
        print("2. Subtraction (-)")
        print("3. Multiplication (*)")
        print("4. Division (/)")
        print("5. Power (^)")
        print("6. Square Root")
        print("7. Percentage (%)")
        print("8. Factorial (!)")
        print("9. Exit")

        choice = input("\nEnter your choice (1-9): ")

        if choice == "9":
            print("\nCalculator Closed. Thank you!")
            break

        try:
            if choice in ["1","2","3","4","5","7"]:
                num1 = float(input("Enter first number: "))
                num2 = float(input("Enter second number: "))

                if choice == "1":
                    result = num1 + num2
                    symbol = "+"

                elif choice == "2":
                    result = num1 - num2
                    symbol = "-"

                elif choice == "3":
                    result = num1 * num2
                    symbol = "*"

                elif choice == "4":
                    if num2 == 0:
                        print("Error: Cannot divide by zero")
                        continue
                    result = num1 / num2
                    symbol = "/"

                elif choice == "5":
                    result = num1 ** num2
                    symbol = "^"

                elif choice == "7":
                    result = (num1 * num2) / 100
                    symbol = "%"

                print(f"\nResult: {num1} {symbol} {num2} = {result}")


            elif choice == "6":
                num = float(input("Enter number: "))

                if num < 0:
                    print("Error: Negative number not allowed")
                    continue

                result = math.sqrt(num)
                print(f"\nSquare Root of {num} = {result}")


            elif choice == "8":
                num = int(input("Enter whole number: "))

                if num < 0:
                    print("Error: Factorial only works on positive numbers")
                    continue

                result = math.factorial(num)
                print(f"\nFactorial of {num} = {result}")


            else:
                print("Invalid Choice! Try again.")

        except ValueError:
            print("Please enter valid numbers.")


# Run Calculator
calculator()
```
